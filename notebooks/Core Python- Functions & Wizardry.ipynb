```python
from urllib.request import urlopen
story = urlopen('http://sixty-north.com/c/t.txt')
story_words = []

for line in story: 
    line_words = line.split()
    for word in line_words:
        story_words.append(word)

story.close()        
story_words #<- story_words all came in byte form as http provides that not str 


#http errors are errors not from python, rather they have to do with the URL or network
```

```python
story = urlopen('http://sixty-north.com/c/t.txt')
story_words = []

for line in story: 
    line_words = line.decode('utf8').split()
    for word in line_words:
        story_words.append(word)
story.close()
story_words

#no longer has lowercase b in front of str, converted using .decode() function
```

```python
# Dictionary Comprehensions- 

#format-> { key_expression(item):value_expression(item) for item in iterable}

country_to_capital = {'United Kingdom':'London',
                     'Brazil':'Brasíia',
                     'Australia':'Canberra',
                     'Japan':'Tokyo'}

capital_to_country = {capital: country for country, capital in country_to_capital.items()}


import pprint as p
p.pprint(capital_to_country)
```

```python
# Classes allow you to create you own 'types' -- defining new types, not the built-in types

# class MyClassName: --> to define new classes 

class Flight:
    
    """typically these include docstrings"""
    
    def __init__(self,number):
        
        if not number[:2].isalpha():
            raise ValueError(f"No airline code in '{number}'")

        if not number[:2].isupper():
            raise ValueError(f"Invalid airline code '{number}'")            

        if not (number[2:].isdigit() and int(number[2::])<= 9999):
            raise ValueError(f"Invalid route code '{number}'")            
            
        self._number = number
        
    def number(self):
        return self._number

```

```python
#implementation of above block

f = Flight('SN60')
f.number()
```
