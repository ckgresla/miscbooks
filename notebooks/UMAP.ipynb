# UMAP

UMAP (Uniform Manifold Approximation and Projection) is a dimensionality reduction/visualization library typically used in Kaggle Comps to get an idea of what classes exist and how they look for a given dataset. 

The normal use case is to call a UMAP Mapper on ML ready data (X, y as two NP Arrays), it comes with sklearn? and is typically used within or around that framework

```python
# Fit UMAP to Dataset

import umap
from sklearn.datasets import load_digits as data #can use load_digits or another dataset from link in above bio

# Sklearn df
dataset = data()

# General UMAP Object Params
mapper = umap.UMAP(n_neighbors=12, #EDIT PARAMS BELOW
                      min_dist=0.0001,
                      metric='correlation').fit_transform(dataset.data) #need change to .fit for plotting
```

```python
# View available params for tweaking (in sep VS Code window)
help(umap.UMAP)
```

```python
# UMAP Object + Params
mapper = umap.UMAP(n_neighbors=12,
                      min_dist=0.00001,
                      metric='correlation').fit(dataset.data)
```

```python
# Scatter Plot 
import umap.plot

#umap.plot.points(mapper, labels=dataset.target, theme="inferno")
#or 
umap.plot.points(mapper, labels=dataset.target, cmap="paired")
```

# Plotting Themes:

- fire

- inferno
- blue
- red
- green

- darkred
- darkgreen

```python
# Connectivity Plot - Without edge_bundling
import umap.plot

umap.plot.connectivity(mapper, labels=dataset.target, show_points=False, theme="blue")
```

```python
# Connectivity Plot - With edge_bundling
import umap.plot

umap.plot.connectivity(mapper, labels=dataset.target, show_points=False, edge_bundling='hammer')
```

```python
# Diagnostic Plot
import umap.plot

umap.plot.diagnostic(mapper, diagnostic_type='vq') #can also use [pca, vq, local_dim, neighborhood] for diagnostic_type
```
