```python
# Read in Toy Dataset
import pandas as pd 

df = pd.read_csv("/Users/ckg-files/Desktop/Programming/Toy Datasets/Meteorite_Landings.csv")
```

```python
# Correlation Matricies
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


plt.figure(figsize=(60,50))
plt.tick_params(labelsize=50, rotation=45);
matrix = np.triu(df.corr())
correlation_mat = df.corr()
sns.heatmap(correlation_mat, annot=False, mask=matrix, vmin=-1, vmax=1, center= 0)
plt.title("Correlation Matrix - All Variables", fontsize=80)
plt.show()
```
