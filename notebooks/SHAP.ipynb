# SHAP 

SHAP (Shapley Additive exPlanations) tries to explain a model through the use of game-theory, answers questions like-
- What features are most important?
- What were the decisions behind any single prediction?

The GitHub is phenomenally well documented, with notebooks and sample code

<BR>

**Links**

SHAP GitHub- https://github.com/slundberg/shap

SHAP Docs- https://shap.readthedocs.io/en/latest/index.html



```python
# GitHub Example

import xgboost
import shap

# train an XGBoost model
X, y = shap.datasets.boston()
model = xgboost.XGBRegressor().fit(X, y)

# Explain with SHAP
# (same syntax works for LightGBM, CatBoost, scikit-learn, transformers, Spark, etc.)
explainer = shap.Explainer(model)
shap_values = explainer(X)

#Visualize the First Prediction

# Waterfall Plot -- feature value & contribution in descending order
shap.plots.waterfall(shap_values[0])

# Force Plot
shap.plots.force(shap_values[0])
```

```python
# All Training Predictions
shap.plots.force(shap_values)
```

```python

shap.plots.scatter(shap_values[:,"RM"], color=shap_values)
```
